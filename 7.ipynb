```python
import numpy as np
from aocd.models import Puzzle
puzzle = Puzzle(year=2021, day=7)
```

```python
data = np.array(puzzle.input_data.split(',')).astype(int)
tst = np.array([16,1,2,0,4,2,7,1,2,14])
```

```python
inp = data
mn = np.min(inp)
mx = np.max(inp)+1
```

```python
r = np.arange(mn,mx)
```

```python
# for each position within upper and lower bounds of positions, calculate fuel
d = [np.sum(np.abs(inp - pos)) for pos in r]
```

```python
# find fuel spent to reach minimum 
d[np.argmin(d)]
```

```python
r = np.arange(450,550)
```

```python
def calc_gas(difs):
    return [np.sum(np.arange(dif)+1) for dif in difs]
```

```python
d = [np.sum(calc_gas((np.abs(inp - pos)))) for pos in r]
```
